```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

n = 2000

df = pd.DataFrame({
    "account_id": range(1, n+1),
    "dpd": np.random.randint(30, 180, n),
    "outstanding_balance": np.random.randint(20000, 500000, n),
    "customer_income": np.random.randint(20000, 150000, n),
    "previous_payments": np.random.randint(0, 10, n),
    "collection_attempts": np.random.randint(0, 6, n),
    "region": np.random.choice(["North","South","East","West"], n),
    "employment_type": np.random.choice(["Salaried","Self-employed","Business"], n)
})
```

```python
df.head()
```

```python
df["recovered_flag"] = (
    (df["dpd"] < 90) &
    (df["previous_payments"] > 3)
).astype(int)
```
